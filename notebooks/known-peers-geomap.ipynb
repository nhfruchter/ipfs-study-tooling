```python
import plotly.plotly as py
import pandas as pd
```

```python
df = pd.read_csv("./unique-peers-geo-sized.csv")
df = df.sort_values(by='occurrences', ascending=False)
```

```python
df.head(n=20)
```

```python
limits = [(0,5),(6,10),(11,20),(21,50),(50,100),(101,1512)]
colors = ["lightgrey", "rgb(237,248,251)","rgb(178,226,226)","rgb(102,194,164)","rgb(44,162,95)","rgb(0,109,44)"][::-1]
scale = 2500
df['text'] = "AS{0} {1} {2}".format(df['asn'], df['aso'], df['occurrences'])

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    city = dict(
        type = 'scattergeo',
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        text = df_sub['asn'],
        marker = dict(
            size = df_sub['occurrences']/scale,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)


```

```python
layout = dict(
        title = 'IPFS AS bubbleplot',
        showlegend = False,
        geo = dict(
            projection=dict( type='Mercator' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

```

```python
import folium
from folium.plugins import HeatMap
```

```python
max_count = df['occurrences'].max()
hmap = folium.Map(location=(0,0), zoom_start=2, tiles='cartodbdark_matter',)
hm = HeatMap(
    list(zip(df.lat.values, df.lon.values, df.occurrences.values)),
    min_opacity=0.2, max_val=max_count, radius=12, blur=12, max_zoom=1
)
```

```python
hmap.add_child(hm)
```

```python
hmap.save("ipfs-client-heatmap.html")
```
